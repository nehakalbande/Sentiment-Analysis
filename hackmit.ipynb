```python
from datasets import load_dataset
dataset = load_dataset("glue","sst2")

dataset
```

```python
import torch
```

```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer


# secret_message = "B FB S NVDA B AAPL S AAPL B TWLO"

checkpoint="HackMIT/double-agent"
model = AutoModelForSequenceClassification.from_pretrained(checkpoint)
tokenizer = AutoTokenizer.from_pretrained(checkpoint)

def tokenize_function(examples):
    return tokenizer(examples["sentence"], padding="max_length", max_length=66,truncation=True)

tokenized_datasets = dataset.map(tokenize_function, batched=True)
# print(tokenized_datasets.f)

# print(tokenized_datasets["validation"])

# tokenized_datasets["validation"].features

# tokenized_datasets/

# inputs = tokenizer(sentences, padding="max_length",max_length=66, truncation=True)
```

```python
small_train_dataset = tokenized_datasets["train"].shuffle(seed=42).select(range(10000))
small_test_dataset = tokenized_datasets["test"].shuffle(seed=42).select(range(1000))
small_val_dataset = tokenized_datasets["validation"].shuffle(seed=42).select(range(800))
 
full_train_dataset = tokenized_datasets["train"]
full_test_dataset = tokenized_datasets["test"]
full_val_dataset = tokenized_datasets["validation"]
```

```python
from transformers import TrainingArguments
from transformers import Trainer

import numpy as np
from datasets import load_metric

metric = load_metric("accuracy")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return metric.compute(predictions=predictions, references=labels)

training_args = TrainingArguments(
    "test_trainer",
    # output_dir='./results',
    per_device_train_batch_size=128,
    per_device_eval_batch_size=128,
    # num_train_epochs=15,
    # learning_rate= 5e-7,
    evaluation_strategy="epoch",
    # save_steps=40000,
    # save_total_limit=10,
    # warmup_steps=500, 
    # weight_decay=0.01,
    )


trainer = Trainer(
    model=model, args=training_args, train_dataset=full_train_dataset, eval_dataset=full_val_dataset,compute_metrics=compute_metrics
    
    
)
trainer.evaluate()
```
